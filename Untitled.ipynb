```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
%matplotlib inline
```

```python
df1=pd.read_csv("data/axial_age.csv")
df2=pd.read_csv("data/social_complexity.csv")
```

```python
no_nulls = set(df1.columns[df1.isnull().mean()==0])
print(no_nulls)
most_missing_cols= set(df1.columns[df1.isnull().mean()>.3333])
print(most_missing_cols)
```

```python
def text_preprocess(str):
    str = str.lstrip('0123456789. ')
    str = re.sub(r"\s+", '_', str)
    str = re.sub(r"[\-.]", '_', str)
    str = str.lower()
    return str

df1.columns = df1.columns.to_series().apply(text_preprocess)
df1["nga"] = df1["nga"].map(text_preprocess)

print("RegExed Columns:\n" + str(df1.columns))
print("\nRegExed NGAs:\n" + str(df1.nga.unique()))
```

```python
df1_subset=df1[["date_from", "nga"]]
df1_subset.describe()
```

```python
df1.date_from.plot(kind="hist");
plt.title("Historical Years Represented")
```

```python
df1.groupby(["nga"]).describe()['date_from']
```

```python
nga_list = df1.nga.unique()
df1_date_list = df1.date_from.unique()
df1_date_list = sorted(df1_date_list)

timespans=[]
y_placement = 0
colormap = plt.cm.gist_ncar
colorst = [colormap(i) for i in np.linspace(0, 0.9, 10)]

for place in nga_list:
    new_dict = {}
    subset = df1[df1["nga"] == place]
    new_dict['nga'] = place
    new_dict['y_placement'] = y_placement
    new_dict['time_start'] = min(subset["date_from"])
    new_dict['time_end'] = max(subset["date_from"])
    new_dict['color'] = colorst[y_placement]
    y_placement += 1
    timespans.append(new_dict)

timedf =pd.DataFrame(timespans)


plt.figure(figsize=(15,5))
plt.hlines(timedf.y_placement, timedf.time_start, timedf.time_end, lw = 10, color = timedf.color)
plt.ylim(-1, len(timedf))
plt.yticks(range(len(timedf)+1), timedf.nga)
plt.title("Time Spans Represented\nin Dataset by Society", fontsize = 20)
plt.xlabel("Years (CE)", fontsize = 15)
plt.show()
```

```python
ckey = 0
plt.figure(figsize=(15,5))
for i in nga_list:
    subset = df1[df1["nga"] == i]
    plt.plot(subset['date_from'],
                    subset['sum'],
                    lw=2, 
             color = colorst[ckey], linestyle = "-.")
    plt.legend(nga_list)
    ckey += 1
plt.title("Total Features Present by Society Over Time", fontsize = 20)
plt.show()
```

```python
df1 = df1.fillna(0)
```

```python
nga_sum = df1.groupby(["nga"]).sum()
nga_sum = nga_sum.drop('date_from', axis = 1)
nga_mean = df1.groupby(["nga"]).mean()
nga_mean = nga_mean.drop('date_from', axis = 1)

fig = plt.figure(figsize=(10,60))
num = 1
for i in nga_sum:
    subdf = nga_sum[[i]]
    ax = fig.add_subplot(13,2,num)
    if i != 'sum':
        subdf.plot.bar(ax = ax, legend = False, title = "{} sum\nby nga".format(i), color = [colorst])
    else:
        subdf.plot.bar(ax = ax, legend = False, title = "total sum by nga", color = [colorst])
    ax.xaxis.label.set_visible(False)
    num += 2

num = 2
for i in nga_mean:
    subdf = nga_mean[[i]]
    ax = fig.add_subplot(13,2,num)
    if i != 'sum':
        subdf.plot.bar(ax = ax, legend = False, title = "{} mean\nby nga".format(i), color = [colorst])
    else:
        subdf.plot.bar(ax = ax, legend = False, title = "mean of total sum\nby nga", color = [colorst])
    ax.xaxis.label.set_visible(False)
    num += 2    

fig.suptitle("Feature Comparison Over Societies", fontsize = 20)
fig.tight_layout(rect=[0, 0.03, 1, 0.97])
fig.show()
```

```python
nga_sum = nga_sum.drop('sum', axis = 1)
nga_sum = nga_sum.transpose()
nga_mean = nga_mean.drop('sum', axis = 1)
nga_mean = nga_mean.transpose()
```

```python
fig = plt.figure(figsize=(10,30))
num = 1
color_idx = 0
for i in nga_mean:
    color = colorst[color_idx]
    subdf = nga_mean[[i]]
    ax = fig.add_subplot(5,2,num)
    subdf.plot.bar(ax = ax, legend = False, title = "total mean of features\nover time in {}".format(i), color = color)
    color_idx += 1
    num += 1
    
fig.suptitle("Features by Society", fontsize = 20)
fig.tight_layout(rect=[0, 0.03, 1, 0.97])
fig.show()
```

```python
df2 = df1.iloc[:,1:14]
corr = df2.corr()
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)

s = corr.unstack()
so = s.sort_values()
plt.title("Correlation", fontsize = 20)
plt.show()
```

```python
col_list = (list(df1))
new_col_list = (col_list[2:-1])

feature_adoption=[]
for i in df1_date_list:
    date_dict = {}
    date_dict["year"] = i
    df_years = df1.loc[df1['date_from'] == i]
    for j in new_col_list:
        df_feature = df_years[["date_from", j]]
        total = df_feature[j].sum()
        date_dict[j] = total
    feature_adoption.append(date_dict)
    
adoptiondf = pd.DataFrame(feature_adoption)



index = 1
for i in range(12):
    adoptiondf.plot(x='year', y=new_col_list[i-1], figsize = (5,2), legend = False)
    plt.title("total {}\nsums over time".format(new_col_list[i-1]))
    plt.ylim(0, 11)
    index += 1
```
